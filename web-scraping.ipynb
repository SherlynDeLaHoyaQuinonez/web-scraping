# INTRODUCCIÓN

El siguiente trabajo es un ejemplo basico de web scraping utilizando la biblioteca requests y parsel en python para extraer informacion de la siguiente pagina web:

https://www.datosmundial.com/america/mexico/crecimiento-poblacional.php

```python
import requests
from parsel import Selector

def scrape_page(url, tag_name):
    # Hacer la petición GET a la página web
    response = requests.get(url)

    # Verificar si la petición fue exitosa (código de estado 200)
    if response.status_code == 200:
        # Obtener el contenido HTML de la página
        html_content = response.text

        # Crear un objeto Selector para analizar el contenido HTML
        selector = Selector(html_content)

        # Encontrar todos los elementos que coinciden con el nombre de la etiqueta
        elements = selector.css(tag_name).getall()

        # Retornar la lista de elementos
        return elements

    else:
        # La petición no fue exitosa, imprimir el código de estado de la respuesta
        print("Error en la petición:", response.status_code)


elements = scrape_page("https://www.datosmundial.com/america/mexico/crecimiento-poblacional.php", "tr")

# Imprimir los elementos encontrados
for element in elements:
    print(element)
```
